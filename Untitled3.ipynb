```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importamos el data set de viajes

df = pd.read_excel("/content/viajes_tup.xlsx")
df
```

```python
#df.groupby(['periodo', 'validaciones_totales']).sum(numeric_only=True)
df['periodo']=pd.to_datetime(df['periodo'])
df['año_viaje'] = df['periodo'].dt.year
```

```python
dtf = df[(df['año_viaje']>=2015) & (df['año_viaje']<=2021)]
```

```python
viajes_por_año = dtf.groupby('año_viaje')['validaciones_totales'].sum()

viajes_año = pd.DataFrame(viajes_por_año)
viajes_año.reset_index(inplace= True)
```

```python
sns.barplot(x='año_viaje',y='validaciones_totales', data=viajes_año, palette= 'deep')
plt.title("Numero de viajes por año")
plt.show()
```

```python
df['fecha'] = pd.to_datetime(df['periodo'])
df['mes'] = df['fecha'].dt.month
df['año'] = df['fecha'].dt.year
data = df.groupby(['mes', 'año'])['validaciones_totales'].sum().reset_index()
```

```python
sns.set(style="darkgrid")
sns.lineplot(x='mes',y='validaciones_totales', hue='año', data=data, palette="bright")
```
